```python
from poliastro.twobody import Orbit
from poliastro.bodies import Earth
from astropy import units as u
import numpy as np
from poliastro.plotting import OrbitPlotter3D
import matplotlib.pyplot as plt
# More info: https://plotly.com/python/renderers/
import plotly.io as pio
pio.renderers.default = "plotly_mimetype+notebook_connected"

# Define the position and velocity vectors for the first orbit
r1 = np.array([7000, 0, 0]) * u.km  # Position vector
v1 = np.array([0, 7, 0]) * u.km / u.s  # Velocity vector

# Define the first orbit
orbit1 = Orbit.from_vectors(Earth, r1, v1)

# Define the position and velocity vectors for the second orbit
r2 = np.array([8000, 50, 65]) * u.km  # Position vector
v2 = np.array([0, 7.5, 0]) * u.km / u.s  # Velocity vector

# Define the second orbit
orbit2 = Orbit.from_vectors(Earth, r2, v2)

# Print the orbits
print("Orbit 1:")
print(orbit1)
print()
print("Orbit 2:")
print(orbit2)
# plt.switch_backend('Qt5Agg')  # Use Qt5 backend
# Create a plotter object
plotter = OrbitPlotter3D()

# Plot the first orbit
hi =plotter.plot(orbit1, label="Orbit 1")
hi.set_markersize(15)
plotter.plot(orbit2, label="Orbit 2")
# plotter.set_markersize(15)
# plotter.ax.scatter(*orbit1.rv()[0], s=10)
# plotter._figure.write_html("satellite_orbits.html",auto_open=True)
# # plotter.savefig('orbit.png')

plotter.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from poliastro.plotting import OrbitPlotter3D
import satelliteSimulation as sim


sim.main()
missDebris=sim.missDebris
orbits=sim.orbits

def plot_orbits(orbit1, orbit2,folderName,i):
    # Create an OrbitPlotter3D instance
    plotter = OrbitPlotter3D()

    # Plot the first orbit
    plotter.plot(orbit1, label="satellite")

    # Plot the second orbit
    plotter.plot(orbit2,label="debris")
    
    plotter._figure.write_html(folderName+"/satellite_orbits"+str(i)+".html")
    plotter.show()

print(orbits)
for name, orbit in orbits.items():
    print(f"{name} orbit:")
    print(orbit)
    print()
    debris=missDebris[name]
    i=0
    for d in debris:
        i=i+1
        plot_orbits(orbit, d,name,i)

        # ss.sureShotSimulation(orbit.r,orbit.v,name)
```
